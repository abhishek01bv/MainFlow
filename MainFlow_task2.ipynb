```python
df1.drop(['Observation'], axis=1, inplace=True) #removing unnecessary column
df1.columns
```

```python
Q1 = df.quantile(0.25) #25 percentile
Q3 = df.quantile(0.75) #75 percentile
```

```python
#inter-quantile method
IQR = Q3-Q1
IQR
```

```python
lower_bound= Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR
```

```python
#removing Outliers using Inter-Quantile method
df1=df1[~(((df1<lower_bound)) | (df1>upper_bound)).any(axis=1)]
```
